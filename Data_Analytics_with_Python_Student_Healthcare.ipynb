**Creating conditional columns**

Step 1: Create the formula (function)  
Step 2: Apply the formula to a new column

```python
# Step 1: Create the formula (function)
def expensive(x):
    if x['negotiated_price'] > 20000:
        return 'Yes'
    else:
        return 'No'
# Step 2: Apply the formula to a new column
cmc['expensive'] = cmc.apply(expensive,axis=1)
```

### Exercise

Make a new column called 'big_difference' that returns 'big' if the price_difference is greater than 50% of the cash_price and 'small' if it is 10% or less

```python
# Step 1: Create the formula (function)
def big_difference(x):
    if x['price_difference'] > x['cash_price'] * 0.5:
        return 'Big'
    elif x['price_difference'] <= x['cash_price'] * .1:
        return 'Small'
    else:
        return 'Medium'
# Step 2: Apply the formula to a new column
cmc['big_difference'] = cmc.apply(big_difference,axis=1)
```

## Merging Datasets

Step 1: Load the other dataframe, look at it

Step 4: Check the number of rows and columns in the original dataframe. Rows should stay the same, columns should be one additional column.

Remember the pt dataframe we loaded earlier? Let's look at the first five rows again

```python
# Step 1: Load the other dataframes, look at it
pt.head()
```

```python
# Step 2: Check the number of rows and columns in the original dataframe
cmc.shape
```

```python
cmc.head()
```

```python
# Step 3: Merge the two dataframes
cmc = pd.merge(cmc,pt,left_on='cpt',right_on='cpt',how='left')
```
